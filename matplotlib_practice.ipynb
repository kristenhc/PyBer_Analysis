```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with object oriented interface ax.plot (versus plt.plot)
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with object oriented interface ax.plt() - different syntax
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with object oriented interface ax.plt() - quick
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# ANNOTATE with MATLAB
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

# Show range of x-axis.
# plt.xlim("Jan", "June")
```

```python
# Annotate with OBJECT ORIENTED

# Create plot.
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label="Boston")
# Add legend
ax.legend()
# Add title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fares($)")
# Add y-axis limit
ax.set_ylim(0,45)
# Add grid lines
ax.grid()

```

```python
# Create vertical bar chart
plt.bar(x_axis, y_axis, color = "green", label="Boston")

# Create labels for x and y axes.
plt.xlabel("Date")
plt.ylabel("PyBer Fare by Month")

# Add a legend
plt.legend()
```

```python
# Create a horizontal bar chart
plt.barh(x_axis, y_axis, color="magenta", label="Boston")
plt.setplt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()

# Invert the y-axis
plt.gca().invert_yaxis()
```

```python
# Create vertical bar chart with object oriented approach.
Fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color="blue")
```

```python
# Create horizonal bar chart w/ object oriented approach.
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
ax.legend()
plt.gca().invert_yaxis()
```

```python
# Create a scatter plot

# Method 1
# plt.plot(x_axis, y_axis, 'o')

#Method 2
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.legend()
plt.gca().invert_yaxis()
```

```python
# Create a bubble chart.
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
# Create larger dots using 
# plt.scatter(x_axis, y_axis, s= y_axis_larger)

# Using list comprehension ' expression for item in list if conditional'
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# Scatter plots - object oriented method
```

```python
# Bubble Chart - object oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='skyblue', s= [i*5 for i in y_axis])
ax.set_xlim(0,40)

plt.gca().invert_yaxis()
```

```python
# Create pie charts MATPLOT

# Add color values
colors = ['slateblue', 'magenta', 'lightblue', 'green', "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Add exploding wedges
explode_values = (0,0,0,0,0,0,0.2,0,0,0,0,0)
# Increase figure size
plt.subplots(figsize=(8,8))

plt.pie(y_axis,
        explode=explode_values,
        colors=colors,
        labels=x_axis,
        autopct='%.1f%%',
       )
plt.title("PyBer Fare by Month")
plt.show()
```

```python
# Create pie chart OBJECT ORIENTED
fig, ax = plt.subplots(figsize=(8,8))
# Add color values
colors = ['slateblue', 'magenta', 'lightblue', 'green', "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
# Add exploding wedges
explode_values = (0,0,.2,0,0,0,0.2,0,0,0,0,0)

ax.pie(y_axis,
       labels=x_axis,
       colors=colors,
       explode=explode_values,
       autopct='%.1f%%',
        startangle= 90,
       shadow = True,
       counterclock = False 
      )
plt.title("PyBer Fare by Month")
plt.show()
```
